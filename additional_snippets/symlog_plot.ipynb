```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
```

```python
import sys
sys.path.append("/home/till/Desktop/GitHub/self-driving-car/src")
from utils import symlog, symexp
```

```python
# Create time steps
x = np.linspace(-20, 20, 200)

# Calculate function values
y_symlog = symlog(x)
y_symexp = symexp(x)
y_id = symexp(symlog(x))

fig, ax = plt.subplots()

ax.plot(x, y_symlog, color="tab:orange", label="symlog")
#ax.plot(x, y_symexp, color="tab:blue", label="symexp")
ax.plot(x, y_id, color="tab:green", label="id ≡ symexp◦symlog ≡ symlog◦symexp")

ax.set_xlabel("Return")
ax.set_ylabel("Scaled Return")

ax.grid(True, alpha=0.1)
ax.legend()
plt.show()
```

```python
x = np.linspace(-12, 12, 200)
y_symlog = symlog(x)
y_id = symexp(symlog(x))

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))  # Create a 1x2 subplot grid

# symlog plot
ax1.plot(x, y_symlog, color="tab:orange", label="symlog")
ax1.plot(x, y_id, color="tab:green", label="id ≡ symexp◦symlog ≡ symlog◦symexp")
ax1.set_xlabel("Return")
ax1.set_ylabel("Scaled Return")
ax1.grid(True, alpha=0.1)
ax1.legend()

# symexp plot
x = np.linspace(-6, 6, 200)
y_symexp = symexp(x)
ax2.plot(x, y_symexp, color="tab:blue", label="symexp")
ax2.set_xlabel("Scaled Return")
ax2.set_ylabel("Return")
ax2.grid(True, alpha=0.1)
ax2.legend()

plt.tight_layout()  # Adjust spacing between subplots
plt.savefig("../figs/symlog_symexp_id_plot.pdf")  
plt.show()
```
